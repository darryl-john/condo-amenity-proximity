```python
import os, glob, json
import pandas as pd
```

```python
records = []

# Append data from all text files in the data_condo_listings directory
for filepath in  glob.glob(os.path.join("data_listings", "*.txt")):
    with open(filepath, 'r', encoding='utf-8') as f:
        for line in f:
            records.append(json.loads(line))

df = pd.DataFrame(records)
df.describe()
```

```python
# Drop 'publish_date' column and remove duplicates
df = df.drop(columns=['publish_date']).drop_duplicates()
df
```

```python
# Save the DataFrame to a CSV file
df.to_csv("condo-listings.csv", index=False, encoding='utf-8-sig')
```
