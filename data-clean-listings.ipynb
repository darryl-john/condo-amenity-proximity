```python
import os, glob, json
import pandas as pd
```

```python
# Open all CSV files in the 'data_listings' directory and append their contents to records
all_dfs = []
for filepath in glob.glob(os.path.join("data_listings", "*.csv")):
    df = pd.read_csv(filepath, encoding='utf-8-sig')
    all_dfs.append(df)

df = pd.concat(all_dfs, ignore_index=True)
df
```

```python
# Drop 'page' and 'publish_date' columns. Remove duplicates and reset index
df = df.drop(columns=['page', 'publish_date']).drop_duplicates().reset_index(drop=True)
df
```
