```python
import os, glob, json, re
import pandas as pd
```

```python
# Open all CSV files in the 'data_listings' directory and append their contents to records
all_dfs = []
for filepath in glob.glob(os.path.join("data_listings", "*.csv")):
    df = pd.read_csv(filepath, encoding='utf-8-sig')
    all_dfs.append(df)

df = pd.concat(all_dfs, ignore_index=True)
df
```

```python
# Drop 'page' and 'publish_date' columns. Remove duplicates and reset index
df = df.drop(columns=['page', 'publish_date']).drop_duplicates().reset_index(drop=True)

# strip multiple spaces in 'title' and 'description' columns
df['title'] = df['title'].str.replace(r'\s{2,}', ' ', regex=True)
df['description'] = df['description'].str.replace(r'\s{2,}', ' ', regex=True)
df
```

```python
df.describe(include='all')
```

```python
# subset columns 'project_name' and 'location'. Drop missing values, remove duplicates and reset index
df_projects = df[['project_name', 'location']].dropna().drop_duplicates().reset_index(drop=True)

# remove string ', Metro Manila' from 'location' column (case-insensitive)
df_projects['location'] = df_projects['location'].str.replace(', Metro Manila', '', case=False).str.strip()

# strip double spaces in projett_name and location columns
df_projects['project_name'] = df_projects['project_name'].str.replace('  ', ' ').str.strip()
df_projects['location'] = df_projects['location'].str.replace('  ', ' ').str.strip()

# group by 'project_name', select 'location' with the longest string length (most specific)
df_projects = df_projects.loc[df_projects.groupby('project_name')['location'].apply(lambda x: x.str.len().idxmax())].reset_index(drop=True)

df_projects
```

```python
# from each row in df, if 'project_name' is empty, find the 1st matching 'project_name' in df_projects that is a substring of 'description' or 'title' and fill in 'project_name' and 'location' from df_projects
for i, row in df.iterrows():
    if pd.isna(row['project_name']) or row['project_name'].strip() == '':
        for j, proj_row in df_projects.iterrows():
            if proj_row['project_name'].lower() in str(row['description']).lower():
                df.at[i, 'project_name'] = proj_row['project_name']
                df.at[i, 'location'] = proj_row['location']
            else:
                if proj_row['project_name'].lower() in str(row['title']).lower():
                    df.at[i, 'project_name'] = proj_row['project_name']
                    df.at[i, 'location'] = proj_row['location']
                    break
```

```python
df.describe(include='all')
```

```python
# remove duplicate rows and filter out rows with missing 'project_name', 'location', 'price', 'floor_area', 'bedrooms', 'bathrooms'
df = df.drop_duplicates().dropna(subset=['project_name', 'location', 'price']).reset_index(drop=True)
df.describe(include='all')
```

```python
# remove non-numeric characters from 'price', 'area', 'floor_area', 'bedrooms', 'bathrooms' columns and convert to numeric dtype
df['price'] = pd.to_numeric(df['price'].str.replace(r'[^\d.]*', '', regex=True), errors='coerce')
df['area'] = pd.to_numeric(df['area'].str.replace(r'[^\d.]*', '', regex=True), errors='coerce')
df['floor_area'] = pd.to_numeric(df['floor_area'].str.replace(r'[^\d.]*', '', regex=True), errors='coerce')
df['bedrooms'] = pd.to_numeric(df['bedrooms'].str.replace(r'[^\d.]*', '', regex=True), errors='coerce')
df['bathrooms'] = pd.to_numeric(df['bathrooms'].str.replace(r'[^\d.]*', '', regex=True), errors='coerce')

df.describe(include='all')
```

```python
# if 'area' <> 'floor_area', set larger value to 'floor_area'. drop 'area' column
df['floor_area'] = df.apply(lambda row: max(row['area'], row['floor_area']) if not pd.isna(row['area']) and not pd.isna(row['floor_area']) else row['floor_area'], axis=1)
df = df.drop(columns=['area'])

# filter out rows with 'price' < 1000000 or 'floor_area' < 18
df = df[(df['price'] >= 1000000) & (df['floor_area'] >= 18)].reset_index(drop=True)
df.describe(include='all')
```

```python
# if any of the strings in ['studio', 'studios', 'studio unit'] is a substring of 'title' or 'description', set 'bedrooms' to 0,
# if any of the strings in ['1 bedroom', '1-bedroom', 'one bedroom', '1br', '1-br', 'one-br'] is a substring of 'title' or 'description', set 'bedrooms' to 1,
# if any of the strings in ['2 bedrooms', '2-bedroom', 'two bedrooms', '2br', '2-br', 'two-br'] is a substring of 'title' or 'description', set 'bedrooms' to 2,
# if any of the strings in ['3 bedrooms', '3-bedroom', 'three bedrooms', '3br', '3-br', 'three-br'] is a substring of 'title' or 'description', set 'bedrooms' to 3

for i, row in df.iterrows():
    title_desc = str(row['title']).lower() + ' ' + str(row['description']).lower()
    if any(x in title_desc for x in ['studio', 'studios', 'studio unit']):
        df.at[i, 'bedrooms'] = 0
    elif any(x in title_desc for x in ['1 bedroom', '1-bedroom', 'one bedroom', '1br', '1-br', 'one-br']):
        df.at[i, 'bedrooms'] = 1
    elif any(x in title_desc for x in ['2 bedrooms', '2-bedroom', 'two bedrooms', '2br', '2-br', 'two-br']):
        df.at[i, 'bedrooms'] = 2
    elif any(x in title_desc for x in ['3 bedrooms', '3-bedroom', 'three bedrooms', '3br', '3-br', 'three-br']):
        df.at[i, 'bedrooms'] = 3
    else:
        df.at[i, 'bedrooms'] = row['bedrooms']
```

```python
# filter out missing values in 'bedrooms' and 'bathrooms' columns
df = df.dropna(subset=['bedrooms', 'bathrooms']).reset_index(drop=True)
df.describe(include='all')
```

```python
# if any of the strings in ['fully furnish', 'fully-furnish'] is a substring of 'title' or 'description', set 'condition' to 'Fully furnished'
# if any of the strings in ['semi furnish', 'semi-furnish', 'partly furnish', 'partly-furnish'] is a substring of 'title' or 'description', set 'condition' to 'Partly furnished'
# if any of the strings in ['unfurnish', 'un-furnish', 'not furnish', 'not-furnish', 'nonfurnish', 'non-furnish'] is a substring of 'title' or 'description', set 'condition' to 'Unfurnished'
for i, row in df.iterrows():
    title_desc = str(row['title']).lower() + ' ' + str(row['description']).lower()
    if any(x in title_desc for x in ['fully furnish', 'fully-furnish']):
        df.at[i, 'condition'] = 'Fully furnished'
    elif any(x in title_desc for x in ['semi furnish', 'semi-furnish', 'partly furnish', 'partly-furnish']):
        df.at[i, 'condition'] = 'Partly furnished'
    elif any(x in title_desc for x in ['unfurnish', 'un-furnish', 'not furnish', 'not-furnish', 'nonfurnish', 'non-furnish']):
        df.at[i, 'condition'] = 'Unfurnished'
```

```python
# filter out rows with missing values in 'condition' column
df = df.dropna(subset=['condition']).reset_index(drop=True)
df.describe(include='all')
```

```python
def find_with_parking(text):
    # Returns the matched string if found, else None
    match = re.search(r'with(\s+\w+)?\s+parking', str(text), re.IGNORECASE)
    if match:
        return match.group(0)
    return None

df['parking_info'] = df.apply(lambda row: find_with_parking(row['title']) or find_with_parking(row['description']), axis=1)

# get numeric parking count from parking_info if available, else set to 1 if parking_info is not null
def extract_parking_count(parking_info):
    if pd.isna(parking_info):
        return 0
    match = re.search(r'(\d+)', parking_info)
    if match:
        return int(match.group(1))
    return 1

# get max between existing 'parking_spaces' column and extracted parking count from parking_info, add new column 'parking'
df['parking'] = df.apply(lambda row: max(row['parking_spaces'], extract_parking_count(row['parking_info'])) if not pd.isna(row['parking_spaces']) else extract_parking_count(row['parking_info']), axis=1)

# drop 'parking_spaces' and 'parking_info' columns
df = df.drop(columns=['parking_spaces', 'parking_info'])
```
