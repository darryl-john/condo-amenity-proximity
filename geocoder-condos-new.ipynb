```python
import os, glob, json, requests
import pandas as pd
```

```python
records = []

# Append data from all text files in the data_condo_listings directory
for filepath in  glob.glob(os.path.join("data_condos_new", "*.txt")):
    with open(filepath, 'r', encoding='utf-8') as f:
        for line in f:
            records.append(json.loads(line))

df = pd.DataFrame(records)
df.describe()
```

```python
# Drop 'publish_date' column and remove duplicates
df = df.drop(columns=['publish_date']).drop_duplicates()
df.describe()
```

```python
# Remove duplicates from 'project_name', 'location', 'floor_area', 'price' columns
df = df.drop_duplicates(subset=['project_name', 'location', 'floor_area', 'price'], keep='first')
df.describe()
```
